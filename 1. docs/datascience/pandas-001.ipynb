```python
# 아래 두개는 동일하다. 무엇을 사용하는지는 상관없지만 일관성있게 사용하자
fare_over_df = df.loc[df['Fare'] >= 30, ['Fare','Name']]
fare_over_df = df.loc[df['Fare'] >= 30][['Fare','Name']]
fare_over_df
```

```python
type(df['Age'])
type(df.loc[df['Age']>20][['Age']])
# 일등석 여성 승객의 평균 나이는
df.loc[(df['Pclass']==1)&(df['Sex']=='female')]['Age'].mean()
```

```python
# Parch가 두명이하인 남성 승객의 평균 연령
df.loc[(df['Parch'] < 2) & (df['Sex'] == 'male')]['Age'].mean()
```

```python
# 파생데이터 생성의 중요성
# df['family_size'] Error 발생
# 1. broadcast로 생성하던가
# 2. 데이터 출력 길이가 같은 것으로 생성하던가
df['family_size'] = 0
df['family_size'] = df['SibSp'] + df['Parch'] + 1
df.head()
```

```python
df['gender_number'] = 0 # 브로드캐스팅 연산
# Warning !
# df.loc[df['Sex'] == 'female'][['gender_number']] = 1 
df.loc[df['Sex'] == 'female', ['gender_number']] = 1
df.loc[df['Sex'] == 'male', ['gender_number']] = 2
df.head()
```
